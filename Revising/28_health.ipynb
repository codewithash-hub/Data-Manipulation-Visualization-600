```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('healthexp.csv')
data.head()
```

```python
# data.loc[data['Country'] == 'Germany']
germany_data = data.loc[(data['Country'] == 'Germany') & (data['Year'] > 1979) & (data['Year'] < 1990)]
germany_data
```

```python
year = pd.Series(germany_data.Year)
spending = pd.Series(germany_data.Spending_USD)
expectancy = pd.Series(germany_data.Life_Expectancy)

plt.title('Germany Total Spending From 1980-1989')
plt.plot(year,spending, color='g', marker='o', linestyle='solid')
plt.ylabel('Spending USD')
plt.xlabel('Years')
plt.show()
```
