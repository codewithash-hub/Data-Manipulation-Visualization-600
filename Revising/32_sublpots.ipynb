```python
import matplotlib.pyplot as plt
```

##### Create a 2x2 grid of subplots, where all axes in the same row share their y-axis scale, and all axes in the same column share their x-axis scale.

```python
# axes are in a 2D array, indexed by [row, (0)]
fig, ax = plt.subplots(2, 4, sharex='col', sharey='row')

for i in range(2):
    for j in range(4):
        ax[i, j].text(0.5, 0.5, str((i, j)), fontsize=18, ha='center')
```
