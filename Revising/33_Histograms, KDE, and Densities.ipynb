```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], density=True, alpha=0.5)
```

Rather than just providing a histogram as a visual output, we can get a smooth estimate
of the distribution using kernel density estimation, which Seaborn does with sns.kdeplot.

```python
sns.kdeplot(data=data, fill=True)
```

If we pass x and y columns to kdeplot, we instead get a two-dimensional visualization
of the joint density

```python
sns.kdeplot(data, x='x', y='y')
```

```python
sns.jointplot(data)
```
