```python
import matplotlib.pyplot as plt
```

```python
movies = ['Avengers', 'Spider-Man', 'Land of Bad', 'Dune part two']
num_oscars = [5, 2, 8, 1]

plt.bar(movies, num_oscars, color='lightblue')
plt.title('Movies & thier oscars')
plt.ylabel('Number of oscars.')
plt.xlabel('Movies')
plt.show()
```
