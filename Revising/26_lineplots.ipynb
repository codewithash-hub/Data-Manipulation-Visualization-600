```python
import matplotlib.pyplot as plt
```

```python
# Data for years and corresponding GDP values
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5976.6, 10289.7, 14958.3]

# creating a line plot with specified attributes
plt.plot(years, gdp, color='g', marker='o', linestyle='solid')
plt.title('Norminal GDP')
plt.ylabel('Billions of Rands.')
plt.xlabel('Years')
plt.show()
```
