Categorical Scatter Plots

```python
# ON Numerical data
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

1. Strip plot

```python
# axes level function 
# On categorical data

sns.stripplot(data=tips, x='day', y='total_bill')
```

```python
# jitter
sns.stripplot(data=tips, x='day', y='total_bill', jitter=False)
```

catplot

```python
# catplot -> Categorical plot
# Figure level functions

sns.catplot(data=tips, x='day', y='total_bill', kind='strip')
```

```python
# hue
sns.catplot(tips, x='day', y='total_bill', kind='strip', hue='sex')
```

2. Swarm Plot

```python
sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
# use by catplot
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex')
```

## Categorical Distribution Plots

#### Boxplot
A boxplot is a standard way of displaying the distribution of data based on a five number summary("minimum", first quartile [Q1], median [Q2], third quartile [Q3], and maximum).

    - It can tell you about outliers and what thier values are.
    - Boxplots can also tell you if your data is symmetrical,
    - how tightly your data is grouped and if and how your data is skewed.

```python
# box plot
sns.boxplot(data=tips, x='day', y='total_bill',)
```

```python
# Using catplot - Figure level
sns.catplot(data=tips, x='day', y='total_bill', kind='box')
```

```python
# hue
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
# color
sns.boxplot(tips, x='day', y='total_bill', hue='sex', palette='dark:gold')
```

```python
# single boxplot -> numerical col
sns.boxplot(tips, y='total_bill', hue='sex', palette='pink')
```

2. Violin Plot

```python
# Violinplot = (Boxplot + KDEplot)
sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
# catplot - figure level function
sns.catplot(data=tips, x='day', y='total_bill', kind='violin')
```

```python
# color
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex', split=True, palette='dark:cyan')
```

```python
# hue
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex', palette='dark:yellow')
```

```python
sns.catplot(tips, x='day', y='total_bill', kind='violin', hue='sex', palette='dark:black', split=True)
```

## Categorical Estimate Plot

1. barplot
- When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using <b>error bars</b>

```python
sns.barplot(tips, x='sex', y='total_bill')
```

```python
# hue
sns.barplot(tips, x='sex', y='total_bill', hue='smoker')
```

```python
# color
sns.barplot(tips, x='sex', y='total_bill', hue='smoker', palette='dark:yellow')
```

```python
# estimate
import numpy as np

sns.barplot(tips, x='sex', y='total_bill', hue='smoker', palette='dark:yellow', estimator=np.std)
```

```python
# estimator (mean)
sns.barplot(tips, x='sex', y='total_bill', hue='smoker', palette='dark:cyan', estimator=np.mean)
```

```python
# estimator (min)
sns.barplot(tips, x='sex', y='total_bill', hue='smoker', palette='dark:cyan', estimator=np.min)
```
