```python
import seaborn as sns
import pandas as pd
```

```python
# Apply the default theme
sns.set_theme()

# Load the data set
data = sns.load_dataset('tips')

# Create a visualization
sns.relplot(data=data, x = "total_bill", y = "tip",
            col = 'time', hue = 'smoker', style = 'smoker', size = 'size')
```

```python
sns.relplot(data=data, x = "total_bill", y = "tip",
            col = 'sex', hue = 'smoker', style = 'smoker', size = 'size')
```
