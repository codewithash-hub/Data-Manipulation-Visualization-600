```python
import matplotlib.pyplot as plt
```

# Line Plots

```python
# Data for years and corresponding GDP values.
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5976.6, 10289.7, 14958.3]

# creating a line chart with specified attributes.
plt.plot(years, gdp, color = 'g', marker = 'o', linestyle = 'solid')

# adding a title to the chart
plt.title("Nominal GDP")

# adding labels to the chart
plt.xlabel("years")
plt.ylabel("Villions of $")

# Displaying the chart
plt.show()

```

# Scatter Plots

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(
        label,
        xy = (friend_count, minute_count),
        xytext = (5, -5),
        textcoords = 'offset points'
)

# Adding title and axis labels
plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("Daily minutes spent on the site")

# Displaying the scatter plot
plt.show()
```
