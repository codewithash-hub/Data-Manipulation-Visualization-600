```python
import matplotlib.pyplot as plt
```

```python
movies = ["Avengers", "Spider-man", "Land of Bad", "Dune part two"]
num_oscars = [5, 2, 8, 1]

plt.bar(movies, num_oscars, color = 'lightblue')
plt.xlabel("Movies")
plt.ylabel("Number of Oscars")
plt.title("Movies and their Awards")
plt.xticks(ticks=movies)
plt.show()
```
