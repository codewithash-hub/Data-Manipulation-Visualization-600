```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

Creating a random set of 10
points on a two-dimensional plane. Using the standard convention, we’ll arrange
these in a 10 × 2 array:

```python
rng = np.random.default_rng(seed=42)
x = rng.random((10, 2))
x
```

To get an idea of how these points look, let’s generate a quick scatter plot

```python
plt.grid()
plt.scatter(x[:, 0], x[:, 1], s=100);
```
